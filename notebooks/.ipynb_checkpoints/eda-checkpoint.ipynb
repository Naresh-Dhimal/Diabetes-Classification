# Exploratory Data Analysis

# Importing important libraries

```python
# for data analysis
import pandas as pd

# for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing dataset

```python
# Specify the absolute dataset path
dataset_path = "../dataset/diabetes.csv"
```

```python
df = pd.read_csv(dataset_path)
df.head(3)
```

# 1. Univariate Analysis:

```python
#checking the columns present in dataset
df.columns
```

### Pregnancies

```python
# to check the distribution of Pregnancies (for numerical features we use .histplot() or .displot())

# Plot the distribution
plt.figure(figsize=(10,10))
sns.displot(data=df["Pregnancies"], kde=True)
plt.title('Distribution of Pregnancies')
plt.xlabel('Pregnancies')
plt.ylabel('Count')
plt.show()
```

#### Based on the distribution plot above for the variable "Pregnancies":
* It can be concluded that the data is right-skewed, indicating that the mean value exceeds the median value.
* The majority of data points fall within the range of 0 to 14 pregnancies.

### Glucose

```python
# to check the distribution of Glucose (for numerical features we use .histplot() or .displot())

# Plot the distribution
plt.figure(figsize=(10,10))
sns.displot(data=df["Glucose"], kde=True)
plt.title('Distribution of Glucose Level')
plt.xlabel('Glucose Level')
plt.ylabel('Count')
plt.show()
```

#### Above Analysis of the Distribution Plot for the Variable "Glucose":
* The distribution plot indicates that the data is approximately normally distributed.
* Most of the data points are within the range of 50 to 200 for glucose levels.

# BloodPressure

```python
# to check the distribution of BloodPressure (for numerical features we use .histplot() or .displot())

# Plot the distribution
plt.figure(figsize=(10,10))
sns.displot(data=df["BloodPressure"], kde=True)
plt.title('Distribution of BloodPressure')
plt.xlabel('BloodPressure')
plt.ylabel('Count')
plt.show()
```

#### Analysis of the Distribution Plot for the Variable "BloodPressure":
* The distribution plot suggests that the data is left-skewed, implying that the median value exceeds the mean value.
* Most of the data points are concentrated within the range of 45 to 110 for blood pressure.

# Age

```python
# to check the distribution of Age (for numerical features we use .histplot() or .displot())

# Plot the distribution
plt.figure(figsize=(10,10))
sns.displot(data=df["Age"], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

#### Analysis of the Distribution Plot for the Variable "Age":

# BMI

```python
# to check the distribution of BMI (for numerical features we use .histplot() or .displot())

# Plot the distribution
plt.figure(figsize=(10,10))
sns.displot(data=df["BMI"], kde=True)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

### Outcome

```python
# to check the distribution of Outcome (for categorical  features we use .barplot() or .pieplot())

# calculating the value counts
count_outcome = df["Outcome"].value_counts()

# Plot the distribution
plt.figure(figsize=(8,8))
sns.barplot(x=count_outcome.index, y=count_outcome.values)
plt.title('Distribution of Outcome')
plt.xlabel('Outcome (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Count')
plt.show()
```

#### Analysis of the Distribution Plot for the Variable "Outcome":
* The distribution plot indicates that the data is imbalanced, particularly for the outcome of diabetes.
* There are approximately 500 instances where individuals do not have diabetes (Outcome = 0) and around 270 instances where individuals have diabetes (Outcome = 1).

# 2. Bivariate Analysis with target "Outcomes"

### Pregnancies vs Outcome

```python
# to check the distribution of numerical vs categorical feature .boxplot() is used 

# Plot the distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='Pregnancies', data=df)
plt.title('Distribution of Pregnancies by Outcome')
plt.xlabel('Outcome (0: No Diabetes, 1: Diabetes)')
plt.ylabel('Number of Pregnancies')
plt.show()
```

### From above distribution,
* The x-axis of the plot represents the outcome categories, with "0" denoting "No Diabetes" and "1" denoting "Diabetes".
* The y-axis represents the number of pregnancies.
* Individuals with diabetes tend to have up to 17 pregnancies, as observed from the plot.
* Individuals with no diabetes tend to have up to 13 pregnancies, as observed from the plot.

### Pregnancies vs Age vs Outcome

```python
# to check the distribution of numerical vs categoricalfeature .scatterplot() is used 

# Plot the distribution
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Pregnancies', hue="Outcome", data=df)
plt.title('Distribution of Pregnancies and Age by Outcome')
plt.xlabel('Age (years)')
plt.ylabel('Pregnancies')
plt.show()
```

### From above distribution,
* As age increases, individuals with diabetes tend to exhibit a greater scatter in the number of pregnancies, as observed from the distribution plot.
* Individuals aged between 30 and 55 who experience an increase in the number of pregnancies are more likely to have diabetes.

# Glucose vs BloodPressure vs outcome

```python
# to check the distribution of numerical vs categoricalfeature .scatterplot() is used 

# Plot the distribution
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BloodPressure', y='Glucose', hue="Outcome", data=df)
plt.title('Distribution of BloodPressure and Glucose by Outcome')
plt.xlabel('BloodPressure (mm Hg)')
plt.ylabel('Glucose Level')
plt.show()
```

### From above distribution,
* Individuals with blood pressure levels between 40 and 115, and glucose levels between 115 and 200, are more likely to have diabetes.


# SkinThickness vs BMI vs outcome


```python
# to check the distribution of numerical vs categoricalfeature .scatterplot() is used 

# Plot the distribution
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='SkinThickness', hue="Outcome", data=df)
plt.title('Distribution of SkinThickness and BMI by Outcome')
plt.xlabel('BMI')
plt.ylabel('SkinThickness')
plt.show()
```

### From above distribution,
* Individuals with a BMI between 25 and 55, and SkinThickness between 15 and 50, exhibit a higher likelihood of having diabetes.
* Additionally, individuals with a BMI of 0 and SkinThickness of 0, as well as those with a BMI of 35 and SkinThickness of 100, also show a propensity for diabetes.



# DiabetesPedigreeFunction vs Insulin vs outcome


```python
# to check the distribution of numerical vs categoricalfeature .scatterplot() is used 

# Plot the distribution
plt.figure(figsize=(8, 6))
sns.scatterplot(x='DiabetesPedigreeFunction', y='Insulin', hue="Outcome", data=df)
plt.title('Distribution of DiabetesPedigreeFunction and Insulin by Outcome')
plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('Insulin')
plt.show()
```

### From above distribution,
* Individuals with a DiabetesPedigreeFunction between 0.2 and 1.5, along with insulin levels between 0 and 300, are more likely to have diabetes.
* Moreover, individuals with a DiabetesPedigreeFunction greater than 1.5 and insulin levels less than 200, as well as those with a DiabetesPedigreeFunction less than 1.5 and insulin levels greater than 400, also demonstrate a predisposition to diabetes.
